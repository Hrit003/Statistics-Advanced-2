[1] **What is hypothesis testing in statistics?**

=> Hypothesis testing in statistics is a method used to make decisions or draw conclusions about a population parameter based on sample data.

It’s a formal process for testing an assumption (the hypothesis) by analyzing evidence from data.

*Example:*

Suppose a company claims their light bulbs last on average 1,000 hours.

* H₀: μ = 1000 (mean life is 1000 hours).

* H₁: μ ≠ 1000 (mean life is different).
You collect a sample, perform a test (say a t-test), calculate a p-value.

* If p ≤ 0.05 → reject H₀ → evidence suggests the average life is not 1000 hours.

* If p > 0.05 → fail to reject H₀ → not enough evidence to say the life differs.

[2] **What is the null hypothesis, and how does it differ from the alternative
hypothesis?**

=>

*Null Hypothesis (H₀):*

* The default assumption in hypothesis testing.

* States that there is no effect, no difference, or no relationship in the population.

* It represents the “status quo” or baseline condition.

* Example: A coin is fair → H₀: p = 0.5 (probability of heads is 0.5).



*Alternative Hypothesis (H₁ or Ha):*

* The opposite claim to the null hypothesis.

* States that there is an effect, a difference, or a relationship.

* It’s what the researcher usually wants to prove.

* Example: The coin is biased → H₁: p ≠ 0.5.

In simple terms:

* H₀: "Nothing new is happening."

* H₁: "Something new/different is happening."


[3] **Explain the significance level in hypothesis testing and its role in deciding
the outcome of a test.**

=>
Definition

* The significance level is the probability of making a Type I error, i.e., rejecting H₀ when it is actually true.

* Common values: 0.05 (5%), 0.01 (1%), or 0.10 (10%).

Role in Hypothesis Testing

1. Set before the test: Researchers choose α before analyzing data.

* Example: α = 0.05 means you accept a 5% risk of wrongly rejecting H₀.

2. Compare with p-value:

* If p ≤ α → reject H₀ (evidence supports the alternative).

* If p > α → fail to reject H₀ (not enough evidence to reject the null).

3. Controls the strictness of the test:

* Smaller α (e.g., 0.01) → more strict, harder to reject H₀, reduces false positives but increases false negatives (Type II error).

* Larger α (e.g., 0.10) → less strict, easier to reject H₀, increases risk of false positives.

[4] **What are Type I and Type II errors? Give examples of each.**

=>
1. Type I Error (False Positive)

* Occurs when you reject the null hypothesis (H₀) even though it is true.

* In other words, you think there’s an effect when there really isn’t.

* Probability of making a Type I error = α (significance level).

Example:
A medical test for a disease:

* H₀: Patient does not have the disease.

* H₁: Patient has the disease.

* If the test says “positive” (disease present) when the patient is actually healthy → Type I error.

2. Type II Error (False Negative)

* Occurs when you fail to reject the null hypothesis (H₀) even though it is false.

* In other words, you miss an effect that actually exists.

* Probability of making a Type II error = β (depends on sample size, variability, effect size).

Example:
Using the same medical test:

* H₀: Patient does not have the disease.

* H₁: Patient has the disease.

* If the test says “negative” (disease absent) when the patient actually has the disease → Type II error.

[5] **What is the difference between a Z-test and a T-test? Explain when to use
each.**

=>

1. Z Test

*Definition*: A statistical test based on the Z-distribution (standard normal distribution).

*When to use:*
* Large sample size (n > 30, Central Limit Theorem ensures normality).
* Population standard deviation (σ) is known.

*Example:*
A factory claims that the average weight of sugar packets is 1 kg. You have a sample of 100 packets, and you know the population standard deviation (σ = 0.05 kg). → Use a Z-test.


2. T Test

*Definition:* A statistical test based on the Student’s t-distribution (heavier tails than normal).

*When to use:*
* Small sample size (n ≤ 30).
* Population standard deviation (σ) is unknown → use sample standard deviation (s) instead.

*Example:*
A teacher wants to test if the average score of 15 students differs from 50, but the population standard deviation is not known. → Use a T-test.

```python
[7] #Implement hypothesis testing using Z-statistics for a sample dataset in Python. Show the Python code and interpret the results.sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
[50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

import numpy as np
from scipy.stats import norm

# Sample data
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

# Hypothesized population mean
mu_0 = 50

# Known population standard deviation
sigma = 1

# Sample statistics
sample_mean = np.mean(sample_data)
n = len(sample_data)

# Z-test statistic
z_stat = (sample_mean - mu_0) / (sigma / np.sqrt(n))

# Two-tailed p-value
p_value = 2 * (1 - norm.cdf(abs(z_stat)))

print("Sample Mean:", sample_mean)
print("Z-statistic:", z_stat)
print("p-value:", p_value)

# Decision at alpha = 0.05
alpha = 0.05
if p_value < alpha:
    print("Reject H0: There is evidence the mean is different from 50.")
else:
    print("Fail to reject H0: No significant evidence the mean differs from 50.")
```

```python
[8] #Write a Python script to simulate data from a normal distribution and calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.

import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

#Step 1: Simulate data
np.random.seed(42)  # for reproducibility
data = np.random.normal(loc=50, scale=5, size=100)  # mean=50, std=5, n=100

#Step 2: Calculate sample mean and standard error
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)  # sample standard deviation
n = len(data)
std_error = sample_std / np.sqrt(n)

#Step 3: 95% confidence interval using t-distribution
confidence_level = 0.95
alpha = 1 - confidence_level
t_critical = stats.t.ppf(1 - alpha/2, df=n-1)  # two-tailed
margin_of_error = t_critical * std_error
ci_lower = sample_mean - margin_of_error
ci_upper = sample_mean + margin_of_error

print(f"Sample Mean: {sample_mean:.2f}")
print(f"95% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")

# Step 4: Plot histogram of data
plt.hist(data, bins=15, edgecolor='black', alpha=0.7, density=True)
plt.axvline(sample_mean, color='red', linestyle='dashed', linewidth=2, label=f'Mean: {sample_mean:.2f}')
plt.axvline(ci_lower, color='green', linestyle='dashed', linewidth=2, label=f'CI Lower: {ci_lower:.2f}')
plt.axvline(ci_upper, color='green', linestyle='dashed', linewidth=2, label=f'CI Upper: {ci_upper:.2f}')
plt.title("Histogram of Simulated Data with Confidence Interval")
plt.xlabel("Value")
plt.ylabel("Density")
plt.legend()
plt.show()
```

```python
[9] # Write a Python function to calculate the Z-scores from a dataset and visualize the standardized data using a histogram. Explain what the Z-scores represent in terms of standard deviations from the mean.


import numpy as np
import matplotlib.pyplot as plt

# Function to calculate Z-scores
def calculate_z_scores(data):
    mean = np.mean(data)
    std = np.std(data, ddof=1)  # sample standard deviation
    z_scores = (data - mean) / std
    return z_scores

# Example dataset
data = np.array([10, 12, 15, 18, 20, 22, 25, 30, 35, 40])

# Calculate Z-scores
z_scores = calculate_z_scores(data)

# Print results
print("Original Data:", data)
print("Z-scores:", np.round(z_scores, 2))

# Plot histogram of Z-scores
plt.hist(z_scores, bins=7, edgecolor='black', alpha=0.7, density=True)
plt.axvline(0, color='red', linestyle='dashed', linewidth=2, label="Mean (0)")
plt.title("Histogram of Standardized Data (Z-scores)")
plt.xlabel("Z-score")
plt.ylabel("Density")
plt.legend()
plt.show()
```
